# Generar archivos de salida
Este notebook crea `files/output/file1.txt`, `file2.txt` y `file3.txt`.

```python
import os
os.makedirs('files/output', exist_ok=True)
for i in range(1, 4):
    src = f'files/input/file{i}.txt'
    dst = f'files/output/file{i}.txt'
    data = ''
    if os.path.exists(src):
        with open(src, 'r', encoding='utf-8') as f:
            data = f.read()
    with open(dst, 'w', encoding='utf-8') as f:
        f.write(data)
print('Archivos generados en files/output')
```
